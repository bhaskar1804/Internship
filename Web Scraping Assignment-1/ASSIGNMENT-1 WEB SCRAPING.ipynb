```python
#2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
#and make data frame

page1= requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')


imdb=BeautifulSoup(page1.content)


Names=[]
for i in imdb.find_all('td',class_="titleColumn"''):
        Names.append(i.text.split()[1:-1])


Year=[]
for i in imdb.find_all('td',class_="titleColumn"''):
        Year.append(i.text.split()[-1])


rating=[]
for i in imdb.find_all('td',class_="ratingColumn imdbRating"''):
        rating.append(i.text.split()[0])


IMDB_TOP_250=pd.DataFrame({'Movie':Names,'Rating':rating,'Year_released':Year})
IMDB_TOP_100=IMDB_TOP_250.iloc[0:100,:]

IMDB_TOP_100
```

```python
#4) Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp . 
meesho= requests.get('https://meesho.com/bags-ladies/pl/p7vbp')


ms=BeautifulSoup(meesho.content,'lxml')
ms

Names=[]
Price=[]
Discount=[]
Discount_amount=[]


for price in ms.find_all('div',class_="Card__BaseCard-sc-b3n78k-0 fVRkfg NewProductCard__PriceRow-sc-j0e7tu-5 dYYUrF NewProductCard__PriceRow-sc-j0e7tu-5 dYYUrF"):
    Price.append(price.h5.text)

for name in ms.find_all('div',class_="Card__BaseCard-sc-b3n78k-0 cXRroa NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 eEBPAI NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 eEBPAI"):
    Names.append(name.p.text)

for dis in ms.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 cZvGTZ"):
    Discount.append(dis.text)

for da in ms.find_all('div',class_="Card__BaseCard-sc-b3n78k-0 iKnziG NewProductCard__DiscountRow-sc-j0e7tu-15 fpGaRM NewProductCard__DiscountRow-sc-j0e7tu-15 fpGaRM"):
    Discount_amount.append(da.text)
    
Product_descr=pd.DataFrame({'Product':Names,'Price₹':Price,'Discount%':Discount,'Discount_amount':Discount_amount})

Product_descr
```

```python
#7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
#and the code for the video from the link for the youtube video from the post.

#7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
#and the code for the video from the link for the youtube video from the post.

cor=requests.get('https://coreyms.com/')
cor

cr=BeautifulSoup(cor.content,'lxml')


Heading = []
Date = []
Content =[]
Code =[]

for i in cr.find_all('article'):
    H = i.h2.a.text
    Heading.append(H)
    
    D =i.find('p',class_="entry-meta").text.split(' ')[:3]
    Date.append(D)
    
    C=i.find('div',class_="entry-content").p.text
    Content.append(C)
    
    try :
        CD=i.find('iframe',class_="youtube-player")['src']
        CD2=CD.split('/')[4]
        CD3=CD2.split('?')[0]
        Code.append(CD3)
    except Exception as e :
        CD=None
        Code.append(CD)


Corey = pd.DataFrame({'Heading': Heading,'Date':Date,'Content':Content,'Youtube_code':Code})
Corey
```

```python
#9) Write a python program to scrape mentioned details from dineout.co.in :

#i) Restaurant name
#ii) Cuisine
#iii) Location
#iv) Ratings
#v) Image URL 

din= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')

DO=BeautifulSoup(din.content)
DO

titles=[]

for i in DO.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)
    
place=[]

for i in DO.find_all('div',class_="restnt-loc ellipsis"):
    place.append(i.text)
    
price=[]

for i in DO.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text[0:13])
    
cuisine=[]

for i in DO.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text[22:])

    
images=[]

for i in DO.find_all('img',class_="no-img"):
    images.append(i['data-src'])
    
ratings=[]

for i in DO.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)

Rest_Details=pd.DataFrame({'Rest_Namae':titles,'Ratings':ratings,'Location':place,'Price':price,'cuisine':cuisine,'images':images})

Rest_Details
```
